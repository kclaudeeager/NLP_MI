```python
def cosineSimillarity(corpus):
    
    start = time.time()
    # Initialize an instance of tf-idf Vectorizer
    tfidf_vectorizer = TfidfVectorizer(stop_words='english')

    # Generate the tf-idf vectors for the corpus
    tfidf_matrix = tfidf_vectorizer.fit_transform(corpus)

    # compute and print the cosine similarity matrix
    #cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
    cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
    #print(cosine_sim)
    # Print time taken
    print("Time taken: %s seconds" % (time.time() - start))
    return cosine_sim
```

```python
#Remove duplicates by loop

def remove_dup(tweetList):
    i = 0
    while i < len(tweetList):
        j = i + 1
        while j < len(tweetList):
            if tweetList[i] == tweetList[j]:
                del tweetList[j]
            else:
                j += 1
        i += 1
    return tweetList

    
```

```python
# print(corpus)
# for text in corpus:
#     print(text)
import re
def removeMentions(text):
    text=re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)","",text)
    return text
```

```python
#adding the replies of the similar tweets  to the tweet in that group
def addRepliesToTweet(similar_group_tweets):
    tweetGroup=similar_group_tweets.get('tweet group')
    orginal_tweet=similar_group_tweets.get('tweet')
 
    replies=orginal_tweet.get('replies')
    for replyObj in tweetGroup:
        newdict=tweetGroup[replyObj]
        #Retturning only the postive replies for each tweet
        bestReplies=[reply for reply  in newdict.get('replies') if json.loads(reply.get("classification")).get('prediction')=='Positive']

        for reply in bestReplies:
            if reply not in replies:
                replyText=reply.get('reply')
                #print(replyText)
                replyText=removeMentions(replyText)
                reply['reply']=replyText
                replies.append(reply)
    orginal_tweet['replies']=replies
    similar_group_tweets['tweet']=orginal_tweet

   
            
#     print(replies)
    
    
```

```python
tweetGroups=getDict_from_json('data\similarTweets_dict_list_file_27_06.json')
for tweet in tweetGroups:
    addRepliesToTweet(tweet)
returnedTweets=[tweet.get('tweet') for tweet in tweetGroups]
print(returnedTweets)
```

```python
#Rearrange the tweets and replies again
def Rearrange_again(returnedTweets):
    for tweet in returnedTweets:
            replies=tweet['replies']
    #         res=json.loads(replies)
    # #         only_replies_texts=res['data']
    #         reply_list=replies
            postive_replies=[rep for rep in replies if (json.loads(rep.get('classification'))).get('prediction')=='Positive']
            negative_replies=[rep for rep in replies if (json.loads(rep.get('classification'))).get('prediction')=='Negative']
            neutral_replies=[rep for rep in replies if (json.loads(rep.get('classification'))).get('prediction')=='Neutral']
            postive_replies=sorted(postive_replies,key=lambda x:x.get('classification'),reverse=True)
            negative_replies=sorted(negative_replies,key=lambda x:x.get('classification'),reverse=False)
            neutral_replies=sorted(neutral_replies,key=lambda x:x.get('classification'),reverse=True)
            reply_list=postive_replies+neutral_replies+negative_replies
            tweet['replies']=reply_list
            #print(tweet)
    return returnedTweets
```
